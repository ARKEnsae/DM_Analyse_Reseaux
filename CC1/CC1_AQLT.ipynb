# 1. ELfriend

```python
ELfriend = nx.read_edgelist('ELfriend_edgelist.txt',comments="node",create_using=nx.DiGraph(),nodetype=int)
print(nx.info(ELfriend))
```

```python
# Il y a 69 nodes plutôt que 71 car il n'y a pas les nodes 44 et 47
# On peut également le vérifier en regardant le fichier .txt
list(set(list(range(1,72)))- set(list(ELfriend.nodes())))
```

```python
# Nodes
ELfriend.nodes()
```

```python
# Edges
ELfriend.edges()
```

```python
ELfriend.out_degree() # out degree
```

```python
ELfriend.in_degree() # in degree
```

```python
neighbors_1 = list(ELfriend.predecessors(1)) + list(ELfriend.successors(1))
# rmq : ici ELadviceG.successors = ELadviceG.neighbors = ELadviceG.adj
print(len(neighbors_1)) # on retrouve 5 = 4+9
print(neighbors_1) # Il y a des doublons : 17 qui est à la fois predecessor et successor
list(dict.fromkeys(neighbors_1)) # récupérer voisins sans doublon
```

```python
## Visualisation

# Paramètre global qui va être utilisé pour le placement des noeuds
global_shell_layout = nx.shell_layout(nx.path_graph(71))
# On crée une fonction pour les graphiques
def graphiques(data, focus_node = 1):
    if type(data).__name__=="DiGraph":
        neighbours = list(data.predecessors(focus_node)) + list(data.successors(focus_node))
    else:
        neighbours = list(data.neighbors(focus_node))
    # On va mettre en évidence les voisins de focus_node (=1)
    node_col = []
    for node in data:
        if node == focus_node:
            node_col.append('red') # 1 en rouge
        elif node in neighbours: 
            node_col.append('orange') # voisins en orange
        else:
            node_col.append('green') # autres en vert
    edge_col = []
    for x,y in data.edges():
        if (x == focus_node)|(y==focus_node):
            edge_col.append('blue') # on met en bleu les arêtes voisins de focus_node
        else: 
            edge_col.append('black') 
    # Pour le draw_shell on veut que le cercle soit ordonné (que les numéros se suivent) :
    pos_shell = nx.shell_layout(data)
    #pos_shell = {sorted(pos_shell.keys())[i]:list(pos_shell.values())[i] for i in range(0,len(pos_shell))}
    pos_shell = {sorted(pos_shell.keys())[i]:global_shell_layout[sorted(pos_shell.keys())[i]-1] for i in range(0,len(pos_shell))}
    
    nx.draw(data, with_labels = True, node_color=node_col, edge_color=edge_col)
    plt.show()
    nx.draw_kamada_kawai(data, with_labels = True, node_color=node_col, edge_color=edge_col)
    plt.show()
    nx.draw(data, with_labels = True, node_color=node_col, edge_color=edge_col, pos = pos_shell)
    plt.show()
```

```python
graphiques(ELfriend) # On met en évidence les voisins du noeud 1
# Les graphiques ne sont pas forcément tous très lisibles
```
